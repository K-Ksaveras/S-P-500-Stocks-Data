# Which sector has the highest total market capitalization?



```python
sector_mc = df.groupby('Sector')['Marketcap'].sum().reset_index().sort_values(by='Marketcap', ascending=False)

sector_mc['Rank'] = range(1, len(sector_mc) + 1)

sector_mc
```

```python
def classification(marketcap):
    if marketcap > 1000000000000:
        return 'Large'
    elif marketcap > 500000000000:
        return 'Medium'
    else:
        return 'Small'

user_value = int(input("Enter a market cap value: "))
print("This company is classified as:", classification(user_value))
```

```python
labels = sector_mc['Sector']
sizes = sector_mc['Marketcap']

plt.pie(sizes, labels=labels, explode=[0.05]*len(sizes), autopct='%.1f %%')
plt.axis('equal')
plt.title('Share of Total Market Cap by Sector')
plt.show()
```

```python
print(df.groupby('State')['ID'].count().sort_values(ascending=False).head(10))
```

```python
largest_marketcap_by_state = df.groupby('State')['Marketcap'].sum().sort_values(ascending=False).head(10)
largest_marketcap_by_state
```

```python
largest_revenue_growth_by_state = df.groupby('State')['Revenuegrowth'].sum().sort_values(ascending=False).head(10)
largest_revenue_growth_by_state
```

```python
states_mc = largest_marketcap_by_state.index
values_mc = largest_marketcap_by_state.values

states_rg = largest_revenue_growth_by_state.index
values_rg = largest_revenue_growth_by_state.values

plt.plot(states_mc, values_mc,
         color='green', linewidth=1, marker='*', linestyle=('-'),
         markersize=10, markeredgecolor='red')
plt.plot(states_rg, values_rg,
         color='green', linewidth=0.5, marker='o', linestyle=("--"),
         markersize=10, markeredgecolor='mediumblue')


plt.title('State Revenue vs State Growth')
plt.xticks(rotation=60)
#plt.legend()
plt.show()
```

# Employees Data

```python
employees = df[['Fulltimeemployees', 'Shortname']]
employees
```

```python
def classification(employee_count):
    if employee_count > 150000:
      return "Big employer company"
    else:
      return "Small/medium employer company"

```

```python
df['EmployerType'] = df['Fulltimeemployees'].apply(classification)
df[['Shortname','Fulltimeemployees','EmployerType']].head(10)
```

```python
value = float(input("Enter the number of employees: "))
print(classification(value))
```

# GalutinÄ—s iÅ¡vados

```python
df['EmployerType'].value_counts().plot.bar(edgecolor="black", linewidth=1.2)

plt.title('Distribution of Employer Types')
plt.xlabel('Employer Type')
plt.ylabel('Number of Companies')
plt.xticks(rotation=360)
plt.show()
```

```python
avg_emp_sector = df.groupby('Sector')['Fulltimeemployees'].mean().sort_values(ascending=False)
avg_emp_sector.plot.bar(color='black', edgecolor='blue', linewidth=1.2)

plt.title('Average Employees per Company by Sector')
plt.xlabel('Sector')
plt.ylabel('Average Employees')
plt.xticks(rotation=90)
plt.show()
```
